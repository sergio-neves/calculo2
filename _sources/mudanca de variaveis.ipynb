## Coordenadas Polares

Considere o problema de calcular uma integral dupla em um domínio como o representado abaixo

```python
# HIDE CODE
t = np.linspace(np.pi/19, np.pi/3,100)

x = 4*np.cos(t)
y = 4*np.sin(t)

x2 = np.cos(np.flip(t))
y2 = np.sin(np.flip(t))

x3 = 4*np.cos(np.pi/19)
y3 = 4*np.sin(np.pi/19)

x = np.append(x,x2)
y = np.append(y,y2)

x = np.append(x,x3)
y = np.append(y,y3)

fig = go.Figure(data = [go.Scatter(x=x, y=y, fill='toself', fillcolor = 'violet')])


fig.update_yaxes(scaleanchor = "x",scaleratio = 1,)

fig.update_xaxes(range=[-0.5,4])
fig.update_yaxes(range=[-0.5,4])

fig.show()
```
